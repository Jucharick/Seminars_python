```python
# Дана функция f(x) = 5*x**2 + 10*x - 30
# 1. Определить корни
# 2. Найти интервалы, на которых функция возрастает
# 3. Найти интервалы, на которых функция убывает
# 4. Построить график
# 5. Вычислить вершину
# 6. Определить промежутки, на которых f > 0 
# 7. Определить промежутки, на которых f < 0
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
limit = 10
step = 0.01
```

```python
a, b, c = 5, 10, -30
```

```python
x = np.arange(-limit, limit, step)
```

```python
def func(x):
    return a*x**2 + b*x + c
```

```python
plt.plot(x, func(x), 'r')
plt.grid()
```

```python
def take_roots(a,b,c):
    discr = b**2 - 4*a*c
    if discr > 0:
        x1 = (-b + discr**0.5)/(2*a)
        x2 = (-b - discr**0.5)/(2*a)
        return x1, x2 # вернется кортеж
    elif discr == 0:
        x = -b/(2*a)
        return x, x # вернется кортеж
    else:
        return None,None
```

```python
roots = take_roots(a,b,c)
```

```python
print(take_roots(5,10,-30))
print(np.roots([5,10,-30])) # используем библиотеку numpy
```

```python
min_y = min(func(x))
print(min_y)
```

```python
# 5*x**2 + 10*x - 30 = y
# 5*x**2 + 10*x - 30 = -35
# 5*x**2 + 10*x + 5 = 0
min_x = take_roots(5, 10, c -min_y)[0]
print(min_x)
```

```python
x_down = np.arange(-limit, min_x, step)
x_up = np.arange(min_x, limit, step)

x_down_pos = np.arange(-limit, min(roots), step)
x_down_neg = np.arange(min(roots), min_x, step)
x_up_neg = np.arange(min_x, max(roots), step)
x_up_pos = np.arange(max(roots), limit, step)
```

```python
plt.rcParams['lines.linestyle'] = '-'
plt.plot(x_down_pos, func(x_down_pos), 'b', label='Убывание, f > 0')
plt.plot(x_up_pos, func(x_up_pos), 'r', label='Возрастание, f < 0')
plt.rcParams['lines.linestyle'] = '--'
plt.plot(x_down_neg, func(x_down_neg), 'b', label='Убывание, f < 0')
plt.plot(x_up_neg, func(x_up_neg), 'r', label='Возрастание, f > 0')

plt.plot(roots[0], func(roots[0]), 'go', label=f'корни {round(min(roots), 2)}, {round(max(roots), 2)}')
plt.plot(roots[1], func(roots[1]), 'go')
plt.plot(min_x, min_y, 'cx', label=f'Экстремум {min_x}, {min_y}')

plt.grid()
plt.legend()
```
